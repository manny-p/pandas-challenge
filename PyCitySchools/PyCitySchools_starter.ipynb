## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python


# calc the following:

# total schools
total_schools = school_data_complete.school_name.nunique()

# total students 
total_students = school_data_complete.student_name.count()

# total budget
total_budget = school_data.budget.sum()

#average math score
average_math_score = school_data_complete.math_score.mean()

#average reading score
average_reading_score = school_data_complete.reading_score.mean()

# % of students with a passing math score (70 or greater)
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, ['student_name']]

# percentage of students with a passing reading score (70 or greater)
passing_reading = school_data_complete.loc[school_data_complete['reading_score'] >= 70, ['student_name']]

# overall passing
overall_passing = school_data_complete.loc[
    (school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70), ['student_name']]

# passing count
passing_math = passing_math.count()
passing_reading = passing_reading.count()
overall_passing = overall_passing.count()

# passing %
percentage_passing_math = int(passing_math) / total_students * 100
percentage_passing_reading = int(passing_reading) / total_students * 100
percentage_overall_passing = int(overall_passing) / total_students * 100

# create df
df_district_summary = pd.DataFrame({
    'Total Schools': [total_schools],
    'Total Students': [total_students],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [percentage_passing_math],
    '% Passing Reading': [percentage_passing_reading],
    '% Overall Passing': [percentage_overall_passing]
})

# format
df_district_summary['Total Budget'] = df_district_summary['Total Budget'].map('${:,.2f}'.format)
df_district_summary['Total Students'] = df_district_summary['Total Students'].map('{:,}'.format)

# show dfS
df_district_summary[[
    'Total Schools',
    'Total Students',
    'Total Budget',
    'Average Math Score',
    'Average Reading Score',
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing',
]]


```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# initialize
school_per_overall = []
school_per_math = []
school_per_reading = []
school_type = []
per_student_budget = []


# format function

def convert_integer(x):
    return int(x)


def convert_currency(x):
    return '${:,.2f}'.format(x)


school_name = school_data_complete['school_name'].unique()

# groupby to calc mean 
grouped_school = school_data_complete.groupby(['school_name'])
df1_school_summary = grouped_school.mean()

# clean data 
df1_school_summary.drop('Student ID', inplace=True, axis=1)
df1_school_summary.drop('School ID', inplace=True, axis=1)

# loop / append data
for school in school_name:
    school_df = school_data_complete[school_data_complete['school_name'] == str(school)]
    school_type_list = school_df['type'].values.tolist()
    school_type.append(school_type_list[0])
    school_budget_list = school_df['budget'].values.tolist()
    per_student_item = float(school_budget_list[0]) / school_df['student_name'].count()
    per_student_budget.append(per_student_item)

    # create df
    overall_passing_df = school_df.loc[
        (school_df['reading_score'] >= 70) & (school_df['math_score'] >= 70), ['student_name']]
    passing_reading_df = school_df.loc[school_df['reading_score'] >= 70, ['student_name']]
    passing_math_df = school_df.loc[school_df['math_score'] >= 70, ['student_name']]

    # passing count
    overall_passing = overall_passing_df.count()
    passing_reading = passing_reading_df.count()
    passing_math = passing_math_df.count()

    # passing %
    school_per_overall.append(int(overall_passing) / school_df['student_name'].count() * 100)
    school_per_reading.append(int(passing_reading) / school_df['student_name'].count() * 100)
    school_per_math.append(int(passing_math) / school_df['student_name'].count() * 100)

# store data
df2_school_summary = pd.DataFrame({
    'school_name': school_name,
    'School Type': school_type,
    'Per Student Budget': per_student_budget,
    '% Passing Math': school_per_math,
    '% Passing Reading': school_per_reading,
    '% Overall Passing': school_per_overall
})

df1_school_summary.rename(columns={
    'reading_score': 'Average Reading Score',
    'math_score': 'Average Math Score',
    'size': 'Total Students',
    'budget': 'Average Budget',
},
    inplace=True)

# merge data
merge_school_summary = pd.merge(df1_school_summary, df2_school_summary, on='school_name', how='outer')

merge_school_summary = merge_school_summary[
    ['school_name', 'School Type', 'Total Students', 'Average Budget', 'Per Student Budget', 'Average Math Score',
     'Average Reading Score', '% Passing Math', '% Passing Reading', '% Overall Passing']]

merge_school_summary.rename(columns={'school_name': 'School Name'}, inplace=True)

copy_school_summary = merge_school_summary.copy()

# format data
merge_school_summary['Total Students'] = merge_school_summary['Total Students'].apply(convert_integer)

merge_school_summary['Average Budget'] = merge_school_summary['Average Budget'].apply(convert_currency)

merge_school_summary['Per Student Budget'] = merge_school_summary['Per Student Budget'].apply(convert_currency)

# sort
merge_school_summary.sort_values(by=['School Name'])

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# format func
def format_value(x):
    return '{:.2f}'.format(x)


# copy data
df_spending = copy_school_summary.copy()

# clean data
df_spending.drop(['Total Students'], inplace=True, axis=1)
df_spending.drop(['Average Budget'], inplace=True, axis=1)

# list by spend
bucket_list = [0, 584.9, 629.9, 644.9, 700]
bucket_list_names = ['<=$585', '$586-630', '$631-645', '$646-680']
df_spending['PPE Range'] = pd.cut(df_spending['Per Student Budget'],
                                  bucket_list, labels=bucket_list_names, include_lowest=True)
# groupby
df_spending_summary = df_spending.groupby('PPE Range').mean()
df_spending_summary.drop(['Per Student Budget'], inplace=True, axis=1)

# format data
df_spending_summary['Average Math Score'] = df_spending_summary['Average Math Score'].apply(format_value)
df_spending_summary['Average Reading Score'] = df_spending_summary['Average Reading Score'].apply(format_value)
df_spending_summary['% Passing Math'] = df_spending_summary['% Passing Math'].apply(format_value)
df_spending_summary['% Passing Reading'] = df_spending_summary['% Passing Reading'].apply(format_value)
df_spending_summary['% Overall Passing'] = df_spending_summary['% Overall Passing'].apply(format_value)

# show data
df_spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python

df_school_size = copy_school_summary.copy()

# clean data

df_school_size.drop(['Per Student Budget'], inplace=True, axis=1)
df_school_size.drop(['Average Budget'], inplace=True, axis=1)

size_list = [0, 1000, 2000, 5000]
group_names = ['Small (<1000)', 'Medium (1000-2000)', 'Large (2000-5000)']

# reset Index 
df_school_size = df_school_size.reset_index()

# spending ranges by list
df_school_size['School Size'] = pd.cut(df_school_size['Total Students'], size_list, labels=group_names)

# groupby range
df_grouped_size = df_school_size.groupby(['School Size'])

df_size_summary = df_grouped_size.mean()

# show data
df_size_summary[[
    'Average Math Score',
    'Average Reading Score',
    '% Passing Math',
    '% Passing Reading',
    '% Overall Passing'
]]
```
